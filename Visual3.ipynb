```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

from pybaseball import lahman
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
standings2025 = pd.read_csv("data/standings2025.csv")
standings2025["Season"] = 2025
standings2024 = pd.read_csv("data/standings2024.csv")
standings2024["Season"] = 2024
standings2023 = pd.read_csv("data/standings2023.csv")
standings2023["Season"] = 2023
standings2022 = pd.read_csv("data/standings2022.csv")
standings2022["Season"] = 2022
standings2021 = pd.read_csv("data/standings2021.csv")
standings2021["Season"] = 2021
standings2020 = pd.read_csv("data/standings2020.csv")
standings2020["Season"] = 2020
standings2019 = pd.read_csv("data/standings2019.csv")
standings2019["Season"] = 2019
standings2018 = pd.read_csv("data/standings2018.csv")
standings2018["Season"] = 2018
standings2017 = pd.read_csv("data/standings2017.csv")
standings2017["Season"] = 2017
standings2016 = pd.read_csv("data/standings2016.csv")
standings2016["Season"] = 2016

teamstats2025 = pd.read_csv("data/teamstats2025.csv")
teamstats2025["Season"] = 2025
teamstats2024 = pd.read_csv("data/teamstats2024.csv")
teamstats2024["Season"] = 2024
teamstats2023 = pd.read_csv("data/teamstats2023.csv")
teamstats2023["Season"] = 2023
teamstats2022 = pd.read_csv("data/teamstats2022.csv")
teamstats2022["Season"] = 2022
teamstats2021 = pd.read_csv("data/teamstats2021.csv")
teamstats2021["Season"] = 2021
teamstats2020 = pd.read_csv("data/teamstats2020.csv")
teamstats2020["Season"] = 2020
teamstats2019 = pd.read_csv("data/teamstats2019.csv")
teamstats2019["Season"] = 2019
teamstats2018 = pd.read_csv("data/teamstats2018.csv")
teamstats2018["Season"] = 2018
teamstats2017 = pd.read_csv("data/teamstats2017.csv")
teamstats2017["Season"] = 2017
teamstats2016 = pd.read_csv("data/teamstats2016.csv")
teamstats2016["Season"] = 2016

standings = pd.concat([standings2025,standings2024,standings2023,standings2022,standings2021,standings2020,standings2019,standings2018,standings2017,standings2016])
teamstats = pd.concat([teamstats2025,teamstats2024,teamstats2023,teamstats2022,teamstats2021,teamstats2020,teamstats2019,teamstats2018,teamstats2017,teamstats2016])
teamstats = pd.merge(standings,teamstats, on = ['Tm','Season'], how = 'left')
teamstats.to_csv("teamstats.csv")
```

```python
teams_1622 = teamstats[(teamstats['Season'] >= 2016) & (teamstats['Season'] <= 2022)]
teams_1622 = teams_1622[['W','L','R_y','SB']]
plot1 = sns.pairplot(teams_1622, diag_kind='kde', kind = "reg", 
             plot_kws={'line_kws':{'color': 'purple'}, 'scatter_kws':{'alpha': 0.2}})
plot1.savefig("Visual3.png")
```

```python
teams_2325 = teamstats[(teamstats['Season'] >= 2023) & (teamstats['Season'] < 2026)]
teams_2325 = teams_2325[['W','L','R_y','SB']]
plot2 = sns.pairplot(teams_2325, diag_kind='kde', kind = "reg", 
             plot_kws={'line_kws':{'color': 'purple'}, 'scatter_kws':{'alpha': 0.2}})
plot2.savefig("Visual4.png")
```
